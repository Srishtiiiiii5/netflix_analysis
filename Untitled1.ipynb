```python
df=pd.read_csv("titles.csv", index_col="title")
```

```python
df.type.unique()
```

### Cleaning data:

```python
df.drop(df[df.release_year < 1998].index, inplace=True)
```

```python
df.id.dropna(inplace=True)
```

```python
df.columns[df.isnull().any()]
```

### Changing to numeric:

```python
df["release_year"]= pd.to_numeric(df.release_year, errors="coerce")
df["runtime"]= pd.to_numeric(df.runtime, errors="coerce")
df["imdb_score"]= pd.to_numeric(df.imdb_score, errors="coerce")
df["imdb_votes"]= pd.to_numeric(df.imdb_votes, errors="coerce")
df["tmdb_popularity"]= pd.to_numeric(df.tmdb_popularity, errors="coerce")
df["tmdb_score"]= pd.to_numeric(df.tmdb_score, errors="coerce")
```

```python
movie_df=df[df.type=="MOVIE"]
```

```python
shows_df= df[df.type=="SHOW"]
```

```python
sns.set_style("darkgrid")
```

### a histogram and count plot which shows the number of movies and shows in each year:

```python
plt.hist([movie_df.release_year, shows_df.release_year], stacked=True)
plt.legend(["movie", "shows"])
plt.title("number of shows and movies in specific years")
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="release_year", hue="type", data=df)
```

```python
movie_and_show= df.type.value_counts()
```

### A piechart which shows number of movies and shows in total

```python
plt.title("movies vs shows")
plt.pie(movie_and_show, labels=movie_and_show.index, labeldistance=1.1, autopct= "%1.1f%%")
```

```python
sns.scatterplot(data=df, x="release_year", y="runtime")
```

```python
plt.plot(df.release_year, df.imdb_score, "o--r")
plt.plot(df.release_year, df.tmdb_score, "x--b", alpha=0.56)
```

```python
plt.hist(df.runtime, color="purple")
plt.xlabel("duration of movies")
plt.ylabel("number of time")
plt.title("maximum number of runtime of movies")
```
